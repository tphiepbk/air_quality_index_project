```python
df_quantrac_215_2022 = df_quantrac[
    (df_quantrac["station"] == 215) &
    ((df_quantrac["date"] >= start_date) & (df_quantrac["date"] <= end_date))
]

df_quantrac_cmaq_215_2022 = df_quantrac_cmaq[
    (df_quantrac_cmaq["station"] == 215) &
    ((df_quantrac_cmaq["date"] >= start_date) & (df_quantrac_cmaq["date"] <= end_date))
]

df_quantrac_cmaq_vientham_215_2022 = df_quantrac_cmaq_vientham[
    (df_quantrac_cmaq_vientham["station"] == 215) &
    ((df_quantrac_cmaq_vientham["date"] >= start_date) & (df_quantrac_cmaq_vientham["date"] <= end_date))
]

df_cmaq_215_2022 = df_cmaq[
    (df_cmaq["station"] == 215) &
    ((df_cmaq["time"] >= start_date) & (df_cmaq["time"] <= end_date))
]

display(df_quantrac_215_2022)
display(df_quantrac_cmaq_215_2022)
display(df_quantrac_cmaq_vientham_215_2022)
display(df_cmaq_215_2022)
```

```python
actual_pm25 = df_quantrac[
    (df_quantrac["station"] == 215) &
    ((df_quantrac["date"] >= start_date) & (df_quantrac["date"] <= end_date))
]["PM2.5"].reset_index(drop=True)

quantrac_pm25 = df_quantrac[
    (df_quantrac["station"] == 215) &
    ((df_quantrac["date"] >= start_date) & (df_quantrac["date"] <= end_date))
]["PM2.5_predicted"].reset_index(drop=True)

quantrac_cmaq_pm25 = df_quantrac_cmaq[
    (df_quantrac_cmaq["station"] == 215) &
    ((df_quantrac_cmaq["date"] >= start_date) & (df_quantrac_cmaq["date"] <= end_date))
]["PM2.5_predicted"].reset_index(drop=True)

quantrac_cmaq_vientham_pm25 = df_quantrac_cmaq_vientham[
    (df_quantrac_cmaq_vientham["station"] == 215) &
    ((df_quantrac_cmaq_vientham["date"] >= start_date) & (df_quantrac_cmaq_vientham["date"] <= end_date))
]["PM2.5_predicted"].reset_index(drop=True)

cmaq_pm25 = df_cmaq[
    (df_cmaq["station"] == 215) &
    ((df_cmaq["time"] >= start_date) & (df_cmaq["time"] <= end_date))
]["pm25"].reset_index(drop=True)

print(actual_pm25)
print(quantrac_pm25)
print(quantrac_cmaq_pm25)
print(quantrac_cmaq_vientham_pm25)
print(cmaq_pm25)
```
